```python
import matplotlib.pyplot as plt

epochs = range(1, 11)
accuracy = [0.8056, 0.8400, 0.8869, 0.8922, 0.8986, 0.9053, 0.9078, 0.9119, 0.9006, 0.8781]
loss = [0.6769, 0.4394, 0.2810, 0.1767, 0.1067, 0.075, 0.0725, 0.0563, 0.0477, 0.0376]

fig, ax1 = plt.subplots()

# Accuracy plot
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Accuracy', color='tab:blue')
ax1.plot(epochs, accuracy, color='tab:blue', marker='o', label='Accuracy')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Loss plot
ax2 = ax1.twinx() 
ax2.set_ylabel('Loss', color='tab:red')
ax2.plot(epochs, loss, color='tab:red', marker='o', label='Loss')
ax2.tick_params(axis='y', labelcolor='tab:red')

# Labels and title
plt.title('XLNet Training Accuracy and Loss')
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Accuracy')
ax2.set_ylabel('Loss')

plt.grid()

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
lines = lines1 + lines2
labels = labels1 + labels2

plt.show()
```
